```python
from pyspark import SparkContext
import json
import sys

sc = SparkContext('local[*]','tip');
data = sc.textFile("yelp_academic_dataset_tip.json").map(lambda x: json.loads(x))
```

Spark Application

```python
data.collect()
```

```python
#generate data of year 2018
tip_cnt = data.filter(lambda x: x["date"][0:4] == '2018').map(lambda a: (a["business_id"],1)).countByKey()
```

```python
result = []
for i in tip_cnt.keys():
    dic={}
    dic["business_id"] = i
    dic["tip_cnt"]= tip_cnt[i]
    result.append(dic)
```

```python
tip_cnt
```

```python
import pandas as pd
result = pd.DataFrame(result)
```

```python
result
```

```python
result.to_csv('tip_cnt.csv')
```
