```python
m = data.pop('stars')
```

```python
#labels encoding
y = []
for i in m:
    if i<3:
        y.append(0)
    elif i<=4:
        y.append(1)
    elif i<=5:
        y.append(2)
```

Upsampling: SMOTE

```python
X_train
```

```python
y_train
```

```python
import copy
df = copy.deepcopy(X_train)
```

```python
df
```

```python
df['stars'] = y_train
```

```python
y_smote = df.pop('categories')
```

```python
x_smote = df
```

```python
x_smote
```

```python
smo = SMOTE(random_state=42)
x_smo, y_smo = smo.fit_resample(x_smote, y_smote)
```

```python
df_smo = pd.concat([x_smo, y_smo],axis = 1)
```

```python
y_train2 = df_smo.pop('stars')
```
